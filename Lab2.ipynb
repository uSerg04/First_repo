Файл “Wind_Solar_Consumption.csv” Построить график разницы потребления и выработки активного потребителя и подсчитать, какое количество электроэнергии он сможет продать во внешнюю систему (или какое ему придется купить) за весь рассматриваемый промежуток времени.

```python
from collections import Counter
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('Wind_Solar_Consumption.csv')
#Второй пункт
print("Заголовок таблицы:")
print(data.columns)
print("Первые несколько строк:")
print(data.head())
print("Последние несколько строк:")
print(data.tail())



df=pd.read_csv('Wind_Solar_Consumption.csv')
df['difference'] = df['Wind_power_MW'] + df['Solar_power_kW']/1000 - df['Consumption_MW']
count = len(df)
time = np.arange(0, count)
plt.plot(time, df['difference'])
plt.xlabel('Время, часы')
plt.ylabel('Разность выработки и потребления, МВт')
plt.title('График почасовой разности выработки и потребления')

sum1 =  df['Wind_power_MW'].sum()
sum2 =  df['Solar_power_kW'].sum()
sum3 =  df['Consumption_MW'].sum()
sum=sum1+sum2/1000+sum3
print('Суммарная выработка за данный период равна ', sum, ' МВт')
```
