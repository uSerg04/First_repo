```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
#Таблица с сайта https://infotables.ru/geografiya/853-krupnejshie-gidroelektrostantsii-rossii
df=pd.read_excel('table.xlsx')
df = df.dropna()
df
```

```python
#сортировка
df=pd.read_excel('table.xlsx')
df = df.dropna()
#Сортировка данных
sorted_df = df.sort_values(by='Гидроэлектростанции России ГЭС', ascending=True)
sorted_df
```

```python
#добавление столбца
mosh = df['Мощность, МВт'].mean()
df1 = df.assign(Отклонение_от_среднего=df['Мощность, МВт'] - mosh)
df1
```

```python
#выборка по условию
condition = df['Мощность, МВт'] > 3500
filtered_df = df[condition]
filtered_df
```

```python
#Выбор фрагмента таблицы по номерам строк и столбцов.
filtered_df = df.iloc[[0, 1, 2], [0, 1]] # Выберите строки с индексами 0, 1 и 2, а также столбцы с индексами 0 и 1
filtered_df
```

```python
#Расчеты статистических показателей по сгруппированным данным.
total_sales_by_category = df.groupby('Река')['Мощность, МВт'].sum()
print(total_sales_by_category)
```

```python
#Построение гистограммы по данным.
plt.bar(df['Гидроэлектростанции России ГЭС'], df['Мощность, МВт'])
plt.xticks(rotation=90)
plt.xlabel('Гидроэлектростанции России ГЭС')
plt.ylabel('Мощность, МВт')
plt.title('Гистограмма мощности ГЭС')

plt.show()
```

```python
#Построение круговой диаграммы по данным.
plt.pie(df['Мощность, МВт'], labels=df['Гидроэлектростанции России ГЭС'])
plt.title('Круговая диаграмма мощности ГЭС')
plt.show()
```

```python
#Построение интерактивного графика по данным.
fig = px.bar(df, x='Гидроэлектростанции России ГЭС', y='Мощность, МВт')
fig.show()
```
